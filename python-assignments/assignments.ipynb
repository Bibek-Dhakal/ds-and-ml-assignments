### Math Tutor Using Random, Arithmetic Operators, and OOP

*Scenario:*
Create a `MathTutor` class that generates random math questions using `random` and `math operators`.

* Track correct and incorrect answers.
* Provide a score at the end.
* Handle invalid input using exception handling.

> Concepts: Random, Math, Exception Handling, Classes

```python
import operator
import random


class MathTutor:
    def __init__(self, num_questions=5):
        self.num_questions = num_questions
        self.correct = 0
        self.incorrect = 0
        # Map string operators to actual math functions
        self.ops = {
            '+': operator.add,
            '-': operator.sub,
            '*': operator.mul,
            '/': operator.floordiv  # Using floor division so answers are always integers
        }

    def generate_question(self):
        num1 = random.randint(1, 20)
        num2 = random.randint(1, 20)
        op_symbol = random.choice(list(self.ops.keys()))

        # Ensure clean division without remainders or zero division
        if op_symbol == '/':
            num1 = num1 * num2

        correct_answer = self.ops[op_symbol](num1, num2)
        return num1, op_symbol, num2, correct_answer

    def start(self):
        print("🎓 Welcome to the Math Tutor! Let's test your skills.")
        print("-" * 45)

        for i in range(self.num_questions):
            num1, op_symbol, num2, correct_answer = self.generate_question()

            # Exception handling for user input
            while True:
                try:
                    user_input = input(f"Q{i + 1}: What is {num1} {op_symbol} {num2}? ")
                    answer = int(user_input)
                    break
                except ValueError:
                    print("⚠️ Invalid input! Please enter a valid integer.")

            if answer == correct_answer:
                print("✅ Correct!\n")
                self.correct += 1
            else:
                print(f"❌ Incorrect. The correct answer was {correct_answer}.\n")
                self.incorrect += 1

        self.show_score()

    def show_score(self):
        print("=" * 45)
        print("📊 Tutor Session Complete!")
        print(f"Total Questions:   {self.num_questions}")
        print(f"Correct Answers:   {self.correct}")
        print(f"Incorrect Answers: {self.incorrect}")

        score_percent = (self.correct / self.num_questions) * 100
        print(f"Final Score:       {score_percent:.2f}%")
        print("=" * 45)


# Run the tutor (you can change the number of questions here)
tutor = MathTutor(num_questions=3)
tutor.start()
```

### Multi-file Project Using Modules and Packages

*Scenario:*
You're building a small finance app.

* Create a package `finance_tools` with modules: `tax.py` and `loan.py`.
* Each module contains utility functions like `calculate_tax()` and `calculate_emi()`.
* Import and use them in a main script that takes user input to do both.

> Concepts: Packages, Modules, Importing, Separation of Concerns


```python
from finance_tools.tax import calculate_tax
from finance_tools.loan import calculate_emi


def finance_app():
    print("💰 Welcome to the Finance App")
    print("-" * 30)
    try:
        # Tax Calculation
        income = float(input("Enter your annual income: $"))
        tax = calculate_tax(income)
        print(f"Calculated Tax (15% default): ${tax:,.2f}\n")

        # Loan Calculation
        principal = float(input("Enter loan principal amount: $"))
        rate = float(input("Enter annual interest rate (in %): "))
        months = int(input("Enter loan tenure in months: "))

        emi = calculate_emi(principal, rate, months)
        print(f"Calculated Monthly EMI: ${emi:,.2f}")

    except ValueError:
        print("⚠️ Invalid input! Please ensure you enter numerical values.")


# Run the app
finance_app()
```

### Exception Handling Scenario: Online Age-Restricted Service

**Scenario:**
You’re building a sign-up system for an online movie rental platform. Some movies are age-restricted (18+). You need to ensure proper validation and error handling during user registration.

**Task:**

1. Create a custom exception class called `UnderageError` that inherits from `Exception`.
2. Write a function `register_user()` that:

   * Takes a user’s `name` and `age` as input.
   * Raises `UnderageError` if the user is under 18.
   * Otherwise, prints a welcome message.
3. Wrap the function call in a `try` block and handle the exception.
4. Use `else` to confirm successful registration and `finally` to always print **“Thank you for using MovieTime!”** regardless of outcome.

Also try to validate if the age input is numeric. Raise a `ValueError` if not, and handle it separately.



```python
# 1. Create custom exception
class UnderageError(Exception):
    """Exception raised for users who are under 18."""
    pass


# 2. Registration function
def register_user(name, age):
    if age < 18:
        raise UnderageError(f"Sorry {name}, you must be at least 18 years old to access age-restricted movies.")
    print(f"🎉 Welcome, {name}! Your registration is successful.")


# 3 & 4. Main execution block with Exception Handling
def sign_up_system():
    print("🎬 Welcome to MovieTime Sign-Up")
    name = input("Enter your name: ")
    age_input = input("Enter your age: ")

    try:
        # Validate if age is numeric (will raise ValueError if not)
        age = int(age_input)

        # Check registration restrictions
        register_user(name, age)

    except ValueError:
        print("❌ Error: Age must be a valid numeric whole number.")
    except UnderageError as e:
        print(f"🚫 Registration Denied: {e}")
    except Exception as e:
        print(f"⚠️ An unexpected error occurred: {e}")
    else:
        # Confirms successful execution with no exceptions
        print("🍿 You can now browse and rent age-restricted movies!")
    finally:
        # Executes regardless of success or failure
        print("Thank you for using MovieTime!\n")


# Run the system
sign_up_system()
```

# Problem Statement

We want to build an online shopping cart system that allows users to add products to their cart, calculate the total cost, apply discounts, and generate an invoice. The system should include the following functionalities:

- Adding products to the cart
- Removing products from the cart
- Calculating the total cost
- Applying discounts based on user type
- Generating an invoice

### 1. Create the Product class

We create a basic `Product` class with attributes for the product name and price.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (${self.price:.2f})"
```

### 2.  Implement the User class

In this step, we create a `User` class with attributes for the user's name and whether they are a premium member.

```python
class User:
    def __init__(self, name, is_premium=False):
        self.name = name
        self.is_premium = is_premium

    def __str__(self):
        status = "Premium" if self.is_premium else "Standard"
        return f"{self.name} [{status} Member]"
```

### 3. Create the ShoppingCart class

In this step, we create a `ShoppingCart` class with methods for adding and removing products from the cart, as well as calculating the total cost of the items in the cart.

**Note**: Define `calculate_total_cost` method in the `ShoppingCart` class, that applies a `10%` discount to the total cost if you are `premium` User.

```python
class ShoppingCart:
    def __init__(self, user):
        self.user = user
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"{product.name} added to cart.")

    def remove_product(self, product_name):
        for product in self.products:
            if product.name == product_name:
                self.products.remove(product)
                print(f"{product.name} removed from cart.")
                return
        print("Product not found in cart.")

    def calculate_total_cost(self):
        total = sum(product.price for product in self.products)

        # Apply 10% discount for premium users
        if self.user.is_premium:
            discount = total * 0.10
            total -= discount

        return total
```

### 4. Testing the functionality

Now that we have implemented the necessary classes and methods, let's test our online shopping cart system:

```python
# Create products
p1 = Product("Laptop", 1000)
p2 = Product("Mouse", 50)
p3 = Product("Keyboard", 80)

# Create users
regular_user = User("Alice")
premium_user = User("Bob", is_premium=True)

# Create shopping cart for premium user
cart = ShoppingCart(premium_user)

# Add products
cart.add_product(p1)
cart.add_product(p2)
cart.add_product(p3)

# Remove a product
cart.remove_product("Mouse")

# Calculate total
total_cost = cart.calculate_total_cost()
print(f"Total cost: ${total_cost}")# Your Solution Here
```

### 5. Generating Invoice for a given cart

```python
def generate_invoice(cart):
    print("\n----- INVOICE -----")
    print(f"Customer: {cart.user.name}")
    print(f"Premium Member: {'Yes' if cart.user.is_premium else 'No'}")
    print("\nItems:")

    total = 0
    for product in cart.products:
        print(f"- {product.name}: ${product.price}")
        total += product.price

    print(f"\nSubtotal: ${total}")

    if cart.user.is_premium:
        discount = total * 0.10
        print(f"Discount (10%): -${discount}")
        total -= discount

    print(f"Total Amount: ${total}")
    print("-------------------")
    
# Generate invoice for the cart
generate_invoice(cart)
```
